# Prostate MRI Dataset

```python
import scipy.io
import numpy as np
import os
import matplotlib.pyplot as plt

dir_path = os.path.join('data', 'data')
```

```python
def trim_zeros_2D(array):

    # trim x
    for i in range(array.shape[0]):
        if np.sum(array[i,:]) > 0:
            break
    for j in reversed(range(array.shape[0])):
        if np.sum(array[j,:]) > 0:
            break
    array = array[i:j+1,:]
    
    # trim y
    for i in range(array.shape[1]):
        if np.sum(array[:,i]) > 0:
            break
    for j in reversed(range(array.shape[1])):
        if np.sum(array[:,j]) > 0:
            break
    array = array[:,i:j+1]
    
    return(array)
```

```python
def trim_zeros_3D(array):

    # trim x
    for i in range(array.shape[0]):
        if np.sum(array[i,:,:]) > 0:
            break
    for j in reversed(range(array.shape[0])):
        if np.sum(array[j,:,:]) > 0:
            break
    array = array[i:j+1,:,:]
    
    # trim y
    for i in range(array.shape[1]):
        if np.sum(array[:,i,:]) > 0:
            break
    for j in reversed(range(array.shape[1])):
        if np.sum(array[:,j,:]) > 0:
            break
    array = array[:,i:j+1,:]
    
    # trim z
    for i in range(array.shape[2]):
        if np.sum(array[:,:,i]) > 0:
            break
    for j in reversed(range(array.shape[2])):
        if np.sum(array[:,:,j]) > 0:
            break
    array = array[:,:,i:j+1]
    
    return(array)
```

```python
data = []
for file in os.listdir(dir_path):
    
    mat_file = scipy.io.loadmat(os.path.join(dir_path, file))
    
    patient_dict = {}
    
    patient_dict['id'] = mat_file['casesTableArr'][0][0][0][0][0][0][0]
    
    patient_dict['T2'] = mat_file['T2']
    
    patient_dict['ADC'] = mat_file['ADC']
    patient_dict['CDI'] = mat_file['CDI']
    patient_dict['HBV'] = mat_file['HBV']
            
    patient_dict['PIRADS_score'] = mat_file['casesTableArr'][0][0][1][0][0]
    patient_dict['curGleason_score'] = mat_file['casesTableArr'][0][0][2][0][0]
    patient_dict['maxGleason_score'] = mat_file['casesTableArr'][0][0][3][0][0]
    
    patient_dict['PIRADS_map'] = mat_file['casesTableArr'][0][0][4]
    patient_dict['curGleason_map'] = mat_file['casesTableArr'][0][0][5]
    patient_dict['maxGleason_map'] = mat_file['casesTableArr'][0][0][6]
    
    patient_dict['mask'] = mat_file['PMask']
    patient_dict['zone_map'] = mat_file['casesTableArr'][0][0][7]
    
    data.append(patient_dict)
```

```python
# Volume-level (ADC/PIRADS)

volume_adc_examples = []
volume_adc_masked_examples = []
volume_adc_extracted_examples = []

volume_adc_pirads_labels = []

for patient in data:
    
    # check if the mask has the same number of slices as the mri volume
    if patient['mask'].shape[-1] != patient['ADC'].shape[-1]:
        continue
    
    binary_mask = patient['mask'] == 255  # create a binary mask
    example = patient['ADC'] * binary_mask  # apply the mask to the volume
    trimmed_example = trim_zeros_3D(example)  # trim the volume to the dimensions of the prostate
    
    volume_adc_examples.append(patient['ADC'])
    volume_adc_masked_examples.append(example)
    volume_adc_extracted_examples.append(trimmed_example)
    
    volume_adc_pirads_labels.append(patient['PIRADS_score'])
```

```python
print(volume_adc_examples[0].shape)
print(volume_adc_masked_examples[0].shape)
print(volume_adc_extracted_examples[0].shape)
```

```python
# Slice-level (ADC/PIRADS)

slice_adc_examples = []
slice_adc_masked_examples = []
slice_adc_extracted_examples = []

slice_adc_pirads_labels = []

for patient in data:
    
    # check if the mask has the same number of slices as the mri volume
    if patient['mask'].shape[-1] != patient['ADC'].shape[-1]:
        continue
    
    for slice_index in range(patient['ADC'].shape[-1]):

        # check the mask to see if the slice contains the prostate
        if np.max(patient['mask'][:,:,slice_index]) > 0:
            
            binary_mask = patient['mask'][:,:,slice_index] == 255  # create a binary mask  
            example = patient['ADC'][:,:,slice_index] * binary_mask  # apply the mask to the slice
            trimmed_example = trim_zeros_2D(example)  # trim the slice to the dimensions of the prostate
            
            slice_adc_examples.append(patient['ADC'][:,:,slice_index])
            slice_adc_masked_examples.append(example)
            slice_adc_extracted_examples.append(trimmed_example)
            
            slice_adc_pirads_labels.append(np.max(patient['PIRADS_map'][slice_index]))
```

```python
plt.imshow(slice_adc_examples[0])
plt.show()

plt.imshow(slice_adc_masked_examples[0])
plt.show()

plt.imshow(slice_adc_extracted_examples[0])
plt.show()
```

```python
# Zone-level (ADC/PIRADS)

zone_adc_masked_examples = []
zone_adc_extracted_examples = []

zone_adc_pirads_labels = []

for patient in data:
    
    # check if the segmentation map has the same number of slices as the mri volume
    if patient['zone_map'].shape[-1] != patient['ADC'].shape[-1]:
        continue
    
    for slice_index in range(patient['ADC'].shape[-1]):
        
        for zone_index in range(10):
            
            # check the zone map to see if the slice contains the zone
            if zone_index + 1 in patient['zone_map'][:,:,slice_index]:
                
                binary_mask = patient['zone_map'][:,:,slice_index] == zone_index + 1  # create a binary mask
                example = patient['ADC'][:,:,slice_index] * binary_mask  # apply the mask to the slice
                trimmed_example = trim_zeros_2D(example)  # trim the slice to the dimensions of the prostate zone
                
                zone_adc_masked_examples.append(example)
                zone_adc_extracted_examples.append(trimmed_example)
                
                zone_adc_pirads_labels.append(patient['PIRADS_map'][slice_index][zone_index])
```
